## K-nearest Neighbors Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data=pd.read_csv("/content/drive/MyDrive/banana_quality.csv",header='infer').values
data[data=='Good']=1
data[data=='Bad']=0
print(data)
X=data[:,0:-1] #input data
y=data[:,-1] #target data
y=y.astype('int')
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y)
#with 80% of the data and predict the class label for the rest 20% of the data

print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
model=KNeighborsClassifier(n_neighbors = 3, weights = 'distance')
model.fit(X_train,y_train)
pred=model.predict(X_test)
accuracy3=accuracy_score(y_test,pred)
print("Accuracy:", accuracy3)
print("Classification Report",classification_report(y_test,pred))
print("Confusion matrix:",confusion_matrix(y_test, pred))
print(len(y_test[y_test!=pred]))
```

```python
model=KNeighborsClassifier(n_neighbors = 5, weights = 'distance')
model.fit(X_train,y_train)
pred=model.predict(X_test)
accuracy5=accuracy_score(y_test,pred)
print("Accuracy:", accuracy5)
print("Classification Report",classification_report(y_test,pred))
print("Confusion matrix:",confusion_matrix(y_test, pred))
print(len(y_test[y_test!=pred]))
```
